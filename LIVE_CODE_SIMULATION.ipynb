```python
# Program Stock Management Supermarket

# Variabel untuk menyimpan data inventory
inventory = []

# Function untuk menambahkan barang ke inventory
def tambah_barang():
    nama = input("Masukkan nama barang: ")
    kategori = input("Masukkan kategori barang: ")
    harga = input("Masukkan harga barang: ")

    # Simpan dalam nested list
    inventory.append([nama, kategori, harga])
    print("Barang berhasil ditambahkan!\n")

# Function untuk menampilkan inventory
def tampilkan_inventory():
    if len(inventory) == 0:
        print("Data belum ada\n")
    else:
        print("\n=== Daftar Inventory ===")
        for i, item in enumerate(inventory, start=1):
            print(f"{i}. Nama: {item[0]}, Kategori: {item[1]}, Harga: {item[2]}")
        print()

# Main program
while True:
    print("=== Sistem Stock Management ===")
    print("1. Tambah Barang")
    print("2. Tampilkan Inventory")
    print("Ketik 'exit' untuk keluar")

    pilihan = input("Pilih menu: ")

    if pilihan == "1":
        tambah_barang()
    elif pilihan == "2":
        tampilkan_inventory()
    elif pilihan.lower() == "exit":
        print("Program selesai. Terima kasih!")
        break
    else:
        print("Pilihan tidak valid, coba lagi.\n")
```

```python
'''
====================================
Graded Challenge 1

Nama    : Ade Widya Ningrum
Batch   : RMT-047

Program ini sebagai program shopping cart sederhana yang memungkinkan user untuk menambah, menghapus dan melihat barang di keranjang belanja
====================================
'''

class CartItem:
    '''
    class disini berfungsi untuk menampung atribut dari data barang yang ada.
    data barang berupa : nama dan harga barang
    nama dan harga ini akan menjadi parameter barang
    '''
    def __init__(self,nama_barang,harga_barang):
        self.nama = nama_barang
        self.harga = harga_barang



class ShoppingCart:
    '''
    class disini berfungsi untuk menampung keranjang belanja user.
    '''
    def __init__(self):
        self.Cart = []
        '''
        karena item yang diinput oleh user akan masuk ke class CartItem
        '''


    def tambah_barang(self):
        '''
        function ini berfungsi untuk menambahkan item barang yang diinput oleh user
        '''
        nama = input("Masukkan nama barang : ")
        harga = int(input("Masukkan harga : "))
        '''
        tipe data harga diubah menjadi int karena akan dilakukan perhitungan total
        '''
        barang = CartItem(nama,harga)
        self.Cart.append(barang)

        print(f"{nama} berhasil ditambahkan")


    def hapus_barang(self):
        nama = input("Masukkan nama barang yang ingin dihapus : ")
        self.Cart = [barang for barang in self.Cart if barang.nama != nama]
        '''
        code ini dipakai untuk mencari item barang pada variabel barang
        yang memiliki nama barang tidak sama dengan nama yang diinput oleh user
        '''
        print(f"{nama} berhasil dihapus di keranjang belanja")


    def tampil_barang(self):
        n = 0
        '''
        karena tabel dimulai dari angka 0
        '''
        for barang in self.Cart:
            print(f"{n}. {barang.nama} - {barang.harga}")
            n +=1
        '''
        fungsi untuk menampilkan list yang ada pada variabel barang mulai dari 0
        dengan perulangan 0+1
        '''

    def lihat_total(self):
        total = 0
        for barang in self.Cart:
            total += barang.harga
        return total
        '''
        memakai return agar selalu terloop
        '''


keranjang = ShoppingCart()
'''
variabel dari class ShoppingCart
'''

print("Selamat Datang di Keranjang Belanja Toko Makmur!")
print()
'''
print ini ditempatkan diatas looping karena hanya akan tampil pada awal program di running'''
while True:
    '''
    memakai while true karena program akan terus berjalan, kecuali user memilih exit
    '''
    print()
    print("Menu :")
    print("1. Menambah Barang")
    print("2. Hapus Barang")
    print("3. Tampilkan Barang di Keranjang")
    print("4. Lihat Total Belanja")
    print("5. Exit")
    print()
    user = input("Pilihan : ")

    if user == "1":
       keranjang.tambah_barang()

    elif user =="2":
        keranjang.hapus_barang()

    elif user =="3":
        keranjang.tampil_barang()

    elif user =="4":
        print(f"Total belanja : {keranjang.lihat_total()}")
        '''
        sesuai instruksi bahwa print dibuat pada interface
        '''

    elif user =="5":
        print("Sampai Jumpa ! Terima kasih sudah belanja di Toko Makmur!")
        break
        '''
        menggunakan break, apabila user memilih exit maka program akan berhenti
        '''

    else :
        print("Pilihannya anda salah. Silahkan coba lagi")
        '''
        error message ketika user menginput tidak sesuai dengan menu yang tersedia
        '''





```

```python
data = []

def input_barang():
    nama = input("Masukkan nama : ")
    harga = int(input("Masukkan harga : "))
    kategori = input("Masukkan kategori: ")

    data.append([nama,harga,kategori])
    print(f"{nama} berhasil disimpan")

def tampil_barang():
    n = 1
    for barang in data:
        nama,harga,kategori = barang
        print(f"{n}. {nama}. Harga : {harga}. Kategori : {kategori}")
        n += 1

while True:
    print("Pilih Menu :")
    print()
    print("1. Input Barang")
    print("2. Tampilkan Barang")
    user = input("Pilihan anda : ")

    if user =="1":
        input_barang()

    elif user =="2":
        tampil_barang()
    else:
        print("Pilihan anda tidak sesuai")

```
