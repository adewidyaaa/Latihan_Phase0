###**SQL**

#####1. Berapa banyak label unique dari bike_type?

````python
# working area

df_bike_type_unique = client.query('''
SELECT COUNT(DISTINCT bike_type)
FROM ftds-hacktiv8-project.practice.bike_rent
''').to_dataframe()

df_bike_type_unique
````

#####2. Apa saja nama label unique dari tipe sepeda?

````python
# working area

df_bike_type_unique2 = client.query('''
SELECT DISTINCT bike_type
FROM ftds-hacktiv8-project.practice.bike_rent
''').to_dataframe()

df_bike_type_unique2
````

#####3. Berapa banyak total peminjaman sepeda pada tahun 2016 dan 2017?

````python
# working area
df_rent_2016_2017 = client.query('''
SELECT
  EXTRACT(YEAR FROM start_time) AS year_time,
  COUNT(*) as total_trip
FROM ftds-hacktiv8-project.practice.bike_rent
WHERE EXTRACT(YEAR FROM start_time) IN (2016,2017)
GROUP BY year_time
''').to_dataframe()

df_rent_2016_2017
````

#####4. Dimana station yang paling banyak dikunjungi sebagai titik keberangkatan pada tahun 2016?

````python
# working area
df_often = client.query('''
SELECT
  start_station_name,
  COUNT(*) as total
FROM ftds-hacktiv8-project.practice.bike_rent
WHERE EXTRACT(YEAR FROM start_time) = 2016
GROUP BY start_station_name
ORDER BY total DESC
LIMIT 1
''').to_dataframe()

df_often
````

#####5. Berapa nilai minimum, maximum, dan rata-rata dari durasi perjalanan pada keseluruhan data?

````python
# working area
df_5 = client.query('''
SELECT
  MAX(duration_minutes) as Max,
  MIN(duration_minutes) as Min,
  AVG(duration_minutes) as Average
FROM ftds-hacktiv8-project.practice.bike_rent
''').to_dataframe()

df_5
````

#####6. Berapa rata-rata durasi perjalanan dari subscriber_type `Student Membership` dan `Local31`?

````python
# working area
df_6 = client.query('''
SELECT
  subscriber_type,
  MAX(duration_minutes) as Max,
  MIN(duration_minutes) as Min,
  AVG(duration_minutes) as Average
FROM ftds-hacktiv8-project.practice.bike_rent
WHERE subscriber_type IN("U.T. Student Membership", "Local30")
GROUP BY subscriber_type
''').to_dataframe()

df_6
````

#####7. Tampilkan entry data yang memiliki `duration_time` diatas 4.25 jam. (*Gunakan subquery untuk menjawab pertanyaan ini dengan meng-convert duration_time dari menit ke hour*).

````python
# working area

df_7 = client.query('''
SELECT *
FROM (
  SELECT *,
  (duration_minutes/60 ) AS duration_time
FROM ftds-hacktiv8-project.practice.bike_rent)
WHERE duration_time > 4.25
''').to_dataframe()

df_7
````
